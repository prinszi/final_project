```python
import pandas as pd
```

```python
inspec_data = "new_inspection_data_2019.csv"
data = pd.read_csv(inspec_data)
data.head()
```

```python
# violation_data = data[data.columns.drop('address', 'aka_name', 'city', 'facility_type', 'inspection_date',
#                                        'inspection_type', 'latitude', 'location', 'longitude',
#                                        'results','risk','state','violations','zip')]

violation_data = data.drop(['address', 'aka_name', 'city', 'dba_name', 'facility_type', 'inspection_date',
                                       'inspection_type', 'latitude', 'location', 'longitude',
                                       'results','risk','state','violations','zip', 'violation_count'], axis = 1)
violation_data.head()
```

```python
violation_type = violation_data[violation_data.columns.drop(list(violation_data.filter(regex='comment')))]
violation_type.head()
```

```python
comment_data = violation_data[violation_data.columns.drop(list(violation_data.filter(regex='type' )))]
comment_data.head()
```

```python
for col in violation_type.columns[2:]:
    print(col)
```

```python
max_number_violations = len(col)
```

```python
# Strip off just the violation number
i = 1
for violation in range(0, max_number_violations):
    viol = f"violation_{i}_type"
    violation_type[viol] = violation_type[viol].str.lstrip().str[:2].str.strip('.')
    i += 1
```

```python
violation_type.head()
```

```python
# Count Violation Types
i = 1
for count in range(0, 63):
    viol_count = f"v{i}"
    viol = f"{i}"
    violation_type[viol_count] = (violation_type.iloc[:,2:] == viol).sum(axis=1)
    i += 1
```

```python
violation_type['violation_count']=data['violation_count']
violation_type.head(10)
```

```python
violation_type['Management']=violation_type[['v1', 'v2', 'v3', 'v4', 'v25', 'v29', 'v32', 'v57', 'v58']].sum(axis=1)
violation_type['Hygienic Practices']=violation_type[['v5', 'v6', 'v7', 'v8', 'v9', 'v10', 'v40']].sum(axis=1)
violation_type['Food Preperation']=violation_type[['v11', 'v12', 'v13', 'v14', 'v15', 'v16', 
                                                    'v17', 'v18', 'v19', 'v20', 'v24', 'v26',
                                                     'v27','v28', 'v30', 'v31', 'v33', 'v34', 
                                                    'v35', 'v36', 'v37', 'v39', 'v42']].sum(axis=1)
violation_type['Pests & Rodents']=violation_type[['v38']].sum(axis=1)
violation_type['Utensils & Equipment']=violation_type[['v41', 'v43', 'v44', 'v45', 'v46', 'v47','v48', 'v49']].sum(axis=1)
violation_type['Phyiscal Facilities']=violation_type[['v50', 'v51', 'v52', 'v53', 'v54', 'v55', 'v56']].sum(axis=1)
violation_type['Compliance']=violation_type[['v59', 'v60', 'v61', 'v62', 'v63']].sum(axis=1)
violation_type.head(20)
```

```python
violation_type.to_csv("violation_count.csv")
```

```python
#Groupby License Number
license_viol_count = violation_type.groupby(['license_']).sum()
```
