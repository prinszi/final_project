```python
#you will need your google API key!!!!!  it is in config under gkey for this file

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from pprint import pprint
from config import gkey
import time
```

```python
#Import City CSV and Check file format
restaurant_df=pd.read_csv("inspection_data_2019.csv")
restaurant_df.head()
```

```python
#****** FINAL TO USE *********
#Prepare Dataframe
places_df = pd.DataFrame([])

i = 0

#Run API Call
for index, row in restaurant_df.iterrows():
    item = row["aka_name"]
    lat = row["latitude"]
    lng = row["longitude"]
    location = f"{lat},{lng}"

    
    try:

        query=item+"&location"+location+"&radius=100"

        # Build URL and Generate Call
        url="https://maps.googleapis.com/maps/api/place/textsearch/json?query="+query+"&key="+gkey
        response_json = requests.get(url).json()
        #print(url)

        #Pull Data from Call and generate lists (**WE CAN ADD STUFF HERE**) 
        name = [item["name"] for item in response_json["results"]]
        rating = [item["rating"] for item in response_json["results"]]
        total_ratings = [item["user_ratings_total"] for item in response_json["results"]]
        address = response_json["results"][0]["formatted_address"]
        glat = response_json["results"][0]["geometry"]["location"]["lat"]
        glng = response_json["results"][0]["geometry"]["location"]["lng"]

        #Generate Initial DataFrame
        places_df = places_df.append(pd.DataFrame({"Data Name": row["aka_name"],
                                                   "Data License": row["license_"], 
                                                   "Data Address": row["address"],
                                                   "Data Lat": row["latitude"],
                                                   "Data Lng": row["longitude"],
                                                   "Google Name":name,
                                                   "User Rating":rating, 
                                                   "Total User Reviews": total_ratings,
                                                   "Google Address": address,
                                                   "Google Lat": glat,
                                                   "Google lng": glng}), ignore_index = True)

    except:
        print("not found " +str(row["aka_name"]))

    continue
```

```python
places_df.head()
```

```python
#REVIEW: I have the condition set to greater than 0 which allows lat or lng to not equal, 
# but we can make it stricter by setting it = 2

# Create column that checks if lat and lngs are equal and drop rows that do not match dataset
places_df['drop_lat'] = np.where(round(places_df["Data Lat"],3)==round(places_df["Google Lat"],3), 1,0)
places_df['drop_lng'] = np.where(round(places_df["Data Lng"],3)==round(places_df["Google lng"],3), 1,0)
places_df['drop_latlng'] = np.where(places_df["drop_lat"] + places_df["drop_lng"] >0, 1,0)
places_df.head()
```

```python
places_df_2 = places_df[places_df['drop_latlng'] == 1]
places_df_2
```

```python
#Develop Data Frame ot the Ratings Summary by City and Interest

#Groupby License Number
places_count_df = places_df.groupby(['Data License'])

#Summary Calculations
average_ratings = round(places_count_df['User Rating'].mean(),2)
average_reviews = round(places_count_df['Total User Reviews'].mean(),2)
total_reviews = places_count_df['Total User Reviews'].sum()
count_places = places_count_df['User Rating'].count()
```

```python
#Final Data Frame
#Match on Data License for Final Merge with City of Chicago Data
google_reviews = pd.DataFrame({'Average of Ratings': average_ratings,
                              'Average Number of Reviews': average_reviews, 
                              'Total Number of Reviews': total_reviews,
                              'Total Returned': count_places,})

google_reviews["Data License"] = google_reviews.index.get_level_values(0)
# google_reviews["interest_sort"] = google_reviews.index.get_level_values(1)

google_reviews.head(20)
```

```python
#**REVIEW**:  20 is the max return from the api call, so i randomly picked 10 as an option to drop on
# seemed like if 20 were being returned, then it is not likely to be the same place under different names

google_reviews = google_reviews[google_reviews['Total Returned'] < 10]
google_reviews
```

```python
#Write Final Dataset to CSV
google_reviews.to_csv("google_reviews.csv", index=False, header=True)
```
